```python
#데이터 입출력
#csv 파일 생성
%%writefile sample1.csv
c1, c2, c3
1, 1, 11, one
2, 2, 22, two
3, 3, 33, three
```

```python
#csv 파일 입력
#csv 파일로부터 읽어 데이터프레임을 만들때는 pandas.read_csv함수를 사용한다
#함수의 입력값으로 파일 이름을 넣는다.
import pandas as pd
pd.read_csv("sample1.csv")
```

```python
#위에서 읽은 데이터에는 열 인덱스는 있지만 행 인덱스 정보가 없으므로
#0부터 시작하는 정수 인덱스가 자동으로 추가되었다.
#만약 데이터 파일에 열 인덱스 정보가 없는 경우에는 read_csv명령의
#names인수로 설정할 수 있다.
%%writefile sample2.csv
1, 1, 11, one
2, 2, 22, two
3, 3, 33, three
```

```python
pd.read_csv('sample2.csv', names=['c1', 'c2'])
```

```python
#csv파일에서 데이터를 구분하는 구분자가 쉼표가 아니면 sep인수를 써서 구분자를 사용자가 지정해준다.
#만약 길이가 정해지지 않은 공백이 구분자인 경우에는 \s+정규식 문자열을 사용한다.
%%writefile sample3.txt
c1       c2        c3        c4
0.179181 -1.538472 1.347553  0.43381
1.024209 0.087307  -1.281997 0.49265
0.417899 -2.002308 0.255245  -1.10515
```

```python
pd.read_table("sample3.txt", sep="\s+")
```

```python
#skuprows : 주석
%%writefile sample4.txt
파일제목 : sample4.txt
데이터 포맷의 설명:
c1, c2, c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
pd.read_csv('sample4.txt', skiprows=[0,1])    #위에서 입력한 문자열 중 한글(1,2번째줄)은 안나온다
```

```python
#na_values : 특정값을 안보이게 하고 싶을때 사용
%%writefile sample5.csv
c1, c2, c3
1, 1.11, one
2, , two
누락, 3.33, three
```

```python
df = pd.read_csv('sample5.csv', na_values=["누락"])
df
```

```python
#csv 파일 출력
#지금까지와 반대로 파이썬의 데이터프레임 값을 csv파일로 출력하고
#싶으면 to_csv메서드를 사용한다.
df.to_csv("sample6.csv")
```

```python
#sep로 구분자를 정할수도 있다
df.to_csv("sample7.txt", sep='|')
```
