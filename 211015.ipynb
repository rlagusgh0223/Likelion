```python
#기상청 현재 날씨 정보 가져오기
#각 도시의 기온과 습도를 가지고와 시각화
import requests
from bs4 import BeautifulSoup
import pandas
import matplotlib as mpl
import matplotlib.pyplot as plt
#기상청 웹사이트 방문.
#기상청 -> 날씨 -> 관측자료 -> 지상관측자료 -> 도시별 현재날씨
source = requests.get('https://www.weather.go.kr/weather/observation/currentweather.jsp')
soup = BeautifulSoup(source.content, 'html.parser')
table = soup.find('table',{'class':'table_develop3'})
data = []
```

```python
for tr in table.find_all('tr'):
    tds = list(tr.find_all('td'))
    for td in tds:
        if td.find('a'):
            point = td.find('a').text
            temp = tds[5].text    #자료중 기온이 있는 줄
            humidity = tds[9].text    #자료중 습도가 있는 줄
            print("{0:<7} {1:<7} {2:<7}".format(point, temp, humidity))#{<7}는 7칸 중 왼쪽정렬 의미
            data.append([point, temp, humidity])
```

```python
#csv 파일로 입력 후 판다스 이용해서 출력
outFp = None
outFp = open("weather.txt", "w", encoding='utf-8')
for tr in table.find_all('tr'):
    tds = list(tr.find_all('td'))
    for td in tds:
        if td.find('a'):
            point = td.find('a').text
            temp = tds[5].text    #자료중 기온이 있는 줄
            humidity = tds[9].text    #자료중 습도가 있는 줄
            outFp.writelines("{0:<7} {1:<7} {2:<7}".format(point, temp, humidity)+'\n')
outFp.close()
```

```python
#모범답안
with open('weather.csv', 'w') as f:  #이게 outFp open이랑 close 한번에 쓰는것과 같다
    f.write("지역, 온도, 습도\n")
    for i in data:
        f.write("{0}, {1}, {2}\n".format(i[0], i[1], i[2]))
    
df = pandas.read_csv('weather.csv', index_col = "지역", encoding='utf-8')#euc-kr이 에러나면 utf-8 쓸것
df
```

```python
#대도시를 중심으로 선택하여 그래프로 표시
#한글 폰트 사용
#legend로 기온과 습도를 지정하여 그래프를 그림
city_df = df.loc[['서울', '인천', '대전', '대구', '광주', '부산', '울산']]
font_name = mpl.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family=font_name)
ax = city_df.plot(kind='bar', title='날씨', figsize=(12, 4), legend=True, fontsize=15)
ax.set_xlabel('도시', fontsize=15)
ax.set_ylabel('기온/습도', fontsize=15)
ax.legend(['기온', '습도'], fontsize=15)
plt.show()
```
