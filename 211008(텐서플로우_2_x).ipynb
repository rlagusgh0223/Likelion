```python
## 여러 입력값을 갖는 로지스틱 회귀 1.x -> 2.x
import tensorflow as tf
import numpy as np

#실행할 때마다 같은 결과를 출력하기 위한 seed값 설정
seed = 0
np.random.seed(seed) #넘파이, 텐서플로우 seed값 초기화
tf.random.set_seed(seed)    #이 부분 1.x랑 다르다

#x, y의 데이터 값
x_data = np.array([[2,3],[4,3],[6,4],[8,6],[10,7],[12,8],[14,19]], dtype=np.float64)#[공부시간, 과외횟수] shape=(7,2)
y_data = np.array([0,0,0,1,1,1,1], dtype=np.float64).reshape(7,1) #shape=(7,1)

#기울기 a와 바이어스 b의 값을 임의로 정함 a의 shape는 2x1
#shpae 2x1의 의미 : 들어오는 값은 2개, 나가는 값은 1개
W = tf.Variable(tf.random.uniform([2,1], dtype=tf.float64))    #shape=(2,1), 여기서는 W를 썼는데 기울기 a이다
b = tf.Variable(tf.random.uniform([1], dtype=tf.float64))    #shape=(1,)

#기울기와 절편을 통해 계산되는 예상 y값
def hyp(W, b):
    #예측치 y = sigmoid(-(Wx+b))
    return tf.sigmoid(tf.matmul(x_data, W) + b)
def costFunc():
    #opt.minimize에서 사용할 손실(비용)함수
    return -tf.reduce_mean(y_data * tf.math.log(hyp(W,b)) + (1 - y_data) * tf.math.log(1 - hyp(W,b)))
def cost(W, b):
    #특정 W와 b에서 손실을 계산하는 함수
    return -tf.reduce_mean(y_data * tf.math.log(hyp(W,b)) + (1 - y_data) * tf.math.log(1 - hyp(W,b)))

#rmse 값을 최소로 하는값 찾기
#확률적 경사하강법
opt = tf.keras.optimizers.SGD(learning_rate = 0.1)

for i in range(3001):    #steps
    opt.minimize(costFunc, var_list=[W, b])#w,b를 최소화하고 그걸로 costFunc에서 사용
    if i%300 == 0:
        print(f'epochs={i}, cost={cost(W,b)}, W1={W.numpy()[0,0]}, W2={W.numpy()[1,0]}, b={b.numpy()}')

print("==================================================")
print("W =",W.numpy())
print("b =", b.numpy())
print("sigmoid =", hyp(W,b).numpy())
print()
print("Accuracy=%.4f" %(sum((y_data == (hyp(W, b).numpy() > 0.5)).flatten()) / len(y_data)))#1과 0 중에서 평균
```
