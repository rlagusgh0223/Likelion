```python
#xerr 인수나 yerr 인수를 지정하면 에러 바를 추가할 수 있다.
#다음 코드에서 alpha는 투명도를 지정한다. 0이면 완전 투명, 1이면 완전 불투명이다.

np.random.seed(0)#랜덤이지만 seed가 0이라 나오는 수가 한정되어있다
#여기서는 0이니까 할때마다 결과가 똑같다
#seed가 달라져야 그때마다 랜덤해서 나오는 결과가 달라진다
#그래서 컴퓨터 타이머를 넣어준다. 타이머는 계속 변경되니까 값이 달라진다.

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)#performance가 바, xerr이 작대기, alpha는 바의 진한 정도
plt.yticks(y_pos, people)
plt.xlabel("x Label")
plt.show()
```

```python
data = [[0.3, 12.27], [-0.78, 14.44], [1.26, 11.87], [0.03, 18.75], [1.11, 17.52], [15.17, 9.29], [0.24, 16.37], [-0.24, 19.78],
       [-0.47, 19.51], [-0.77, 12.65], [-0.37, 14.74], [-0.85, 10.72], [-0.41, 21.94], [-0.27, 12.83], [0.02, 15.51], [-0.76, 17.14],
       [2.66, 14.42]]

#del data[5]

x_data = [i[0] for i in data]
y_data = [i[1] for i in data]

del x_data[5]#극단치(5번째 값) 제거
del y_data[5]

import tensorflow.compat.v1 as tf
tf.compat.v1.disable_v2_behavior()
import matplotlib.pyplot as plt
import numpy as np

a = tf.Variable(tf.random_uniform([1], 0, 10, dtype=tf.float64, seed=0))
b = tf.Variable(tf.random_uniform([1], 0, 100, dtype=tf.float64, seed=0))

y = a * x_data + b
learning_rate = 0.1
rmse = tf.sqrt(tf.reduce_mean(tf.square(y - y_data)))
gradient_decent = tf.train.GradientDescentOptimizer(learning_rate).minimize(rmse)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for i in range(2001):
        sess.run(gradient_decent)
        
        if i%100 == 0:
            print("epoch = %.4f, rmse = %.4f, 기울기 a = %.4f, 절편 b = %.4f"%(i, sess.run(rmse), sess.run(a), sess.run(b)))
            
            
    data_a = sess.run(a)    #최종 구해진 기울기와 절편 입력
    data_b = sess.run(b)

line_x= np.arange(min(x_data), max(x_data), 0.01)#증가의 작은값, 큰값을 0.01단위로 구분
line_y = data_a * line_x + data_b
#print(line_x)
#print(line_y)
            
#위의 결과를 그래프로
#plt.figure(figsize=(10,7))
plt.plot(line_x, line_y, 'r-')    #선
plt.plot(x_data, y_data, 'bo')    #데이터
plt.xlabel("Population Growth Rate(%)")
plt.ylabel("Elderly Growth Rate(%)")
#plt.xticks(np.arange(-1, 3, 1))
plt.yticks(np.arange(-30, 50, 10))
plt.show()
```
